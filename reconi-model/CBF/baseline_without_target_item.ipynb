```python
import numpy as np
import pandas as pd

from sklearn.metrics.pairwise import cosine_similarity
```

```python
beans_data = pd.read_csv('../data/원두정보메타데이터.csv')
print(beans_data.shape)

merged_data = pd.read_csv('../data/merged_data.csv') # 콩스콩스만 리뷰가 있으므로 콩스콩스 원두 데이터만 포함
print(merged_data.shape)
```

```python
# 각 유저에 대한 아이템 구매 이력 리스트 생성
item_set_for_each_user = {}
for user_name in merged_data['구매자 이름'].unique(): 
    item_set_for_each_user[user_name] = set(merged_data[merged_data['구매자 이름'] == user_name]['상품명'])

print(item_set_for_each_user)
```

```python
# 병합을 위해 상품명 통일
beans_data = beans_data.rename(columns={"Bean name": "상품명"})
filter_index = beans_data[beans_data["로스터리"] == "콩스콩스"].index
beans_data.loc[filter_index, "상품명"] = beans_data.loc[filter_index, "상품명"].apply(lambda x: x.replace(" ", "")).apply(lambda x: x.lower())
beans_data
```

```python
features_to_use = [
    # "Cupping Note 향미",
    "Cupping Note 산미",
    "Cupping Note 단맛",
    "Cupping Note 바디감",
    # "Rosting Point",
    # "리뷰 수",
]

user_profile_list = []
user_name_list = []

for user_name, item_set in item_set_for_each_user.items():
    items_for_each_user = beans_data[
        (beans_data["상품명"].isin(item_set)) & (beans_data["로스터리"] == "콩스콩스")
    ][features_to_use]
    
    user_profile = items_for_each_user.mean().to_frame().T
    user_profile_list.append(user_profile)
    user_name_list.append(user_name)

user_profile = pd.concat(user_profile_list, ignore_index=True)
user_profile.insert(0, '사용자 이름', user_name_list)
```

```python
item_profile= beans_data[features_to_use]
item_profile
```

```python
# 코사인 유사도 계산
cosine_sim = cosine_similarity(item_profile, user_profile)

cosine_sim_df = pd.DataFrame(
    cosine_sim,
    index=beans_data['상품명'], # item id
    columns=user_name_list, # user id
)
```
